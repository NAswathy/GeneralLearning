shallow copy is same as copy.copy(). This is the second method of "copying a list". While the first method is to just assign a list to a new variable. In this case, the new variable is just a reference to the old list. So, if you change the new list, the old list will also change and vice versa. The shallow copy is similar to the first method ie assignment.In shallow copy, just like te assignment type, creates a new object and fills with the references of the original copy. Normal changes like adding removing etc of the original list won't affect the other copy in a shallow list. But if changes are made to the inner lists etc, in the shallow copy or the original copy, then the changes done for the first list will be reflected in the second list as well, viceversa.

shallow copy (copy.copy) contradicts with the name, the changes in the inner lists gets reflected in the original one. But not the changes in the actual list. In hard copy or deep copy, it is quite ""hard" in handling, ie any change in the deep cpy will not affect the original copy

```python
import copy
original = [1, 2, [3, 4]]
original_assign=original
print(original,original_assign)
shallow_copy = copy.copy(original)
print(original,original_assign)
shallow_copy[2][0] = 5
print(original,original_assign, shallow_copy)  # Outputs: [1, 2, [5, 4]]
```

```python
import copy
original = [1, 2, [3, 4]]
deep_copy = copy.deepcopy(original)
print(deep_copy, original)
deep_copy[2][0]=25
deep_copy.append(50)
print(deep_copy, original)
```

in deep copy, the elements are recursively added to the new variable and not the references to the original element. Hence both the copies remain intact.
