```python
a=10
print("Local scope a is {}" .format(a))
print(f"local scope is {a}")
print (a )
def fun():
    global a
    a=20
    
fun()
print(a)
```

```python
def outer():
    variable = "hello 1" # This variable is declared in a Local scope
    
    def inner():
        nonlocal variable # Without this keyword we could not change the variable value
        variable = "hello 2"
    
    inner()
    print(variable)
    
outer() # "hello 2"
#inner() #inner can't be called as it is defined inside the outer function
```
