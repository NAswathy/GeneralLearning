```python
def my_decorator(original_function):
    def wrapper_function(*args, **kwargs):
        print("Before original function")
        result = original_function(*args, **kwargs)
        print("After original function")
        return result
    return wrapper_function

@my_decorator
def display():
    print("Inside original function")

display()
```

```python
import time 
  
  
def timeis(func): 
    '''Decorator that reports the execution time.'''
    print("inside timeis")
  
    def wrap(*args, **kwargs): 
        start = time.time() 
        result = func(*args, **kwargs) 
        end = time.time() 
          
        print(func.__name__, end-start, args[0]) 
        return result 
    print ("outside wrap")
    return wrap 
  
@timeis
def countdown(n): 
    '''Counts down'''
    while n > 0: 
        n -= 1
  
countdown(5) 
countdown(1000) 
```
