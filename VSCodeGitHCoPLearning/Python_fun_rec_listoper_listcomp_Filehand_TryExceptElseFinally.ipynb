```python
def permute(nums):
    if len(nums) <= 1:
        return [nums]
    
    result = []
    for i, num in enumerate(nums):
        remaining_nums = nums[:i] + nums[i + 1:]
        for perm in permute(remaining_nums):
            result.append([num] + perm)
    
    return result

# Example usage:
array = [1, 2, 3]  # Replace this with your array of integers
result = permute(array)
print(result)
print(len(result))
```

```python
#Recursive factorial
def fact(n):
    if n ==1 or  n==0:
        return 1
    else:
        return (n*fact(n-1))
    
n=5    
print("Factorial of %d is %d" % (n,fact(n)))
```

```python
def checkprime(n):
    if n<0 :
        return (str(n)+" is a negative number")
    if not n<0 and (n==1 or n==0):
        return (str(n)+" is neither prime not composite")
    elif n>1:
         cnt=0
         for i in range(1,n//2):
                if n%i ==0:
                    cnt+=1
                
         if cnt==1:
                return (str(n)+" is a prime number")
         elif cnt>1:
                return (str(n)+" is not a prime number")
            
            
num=21
print (checkprime(num))
```

```python
#Min and Max from a list
def maxmin(l):
    if len(l)==1:
        M=l[0]
        m=l[0]
        return M,m
    else:
        M=l[0]
        m=l[0]
        for i in l:
            if i> M:
                M=i
            if i<m:
                m=i
                
        return M,m
    
lists=[4,4,6,100,2,-105, -105]

print("The max number is %d and min number is %d " %(maxmin(lists)))
```

```python
import re

# Sample text with patterns to replace
text = "apple123ABCd banana45 cherry7cABC date999TrfA"

# Pattern: one or more letters followed by one or more digits
pattern = r'([a-zA-Z]+)(\d+)([A-Za-z]+)' #r'([a-zA-Z]+)(\d+)'

# Replace with a custom format-using "re" library substitute this pattern by this text or another pattern in this string
# This example will put the number first, followed by the text in uppercase
result = re.sub(pattern, r'\2-\1_REPLACED\3', text)
result=re.sub(pattern, r'\2-\1_REPLACED-\3',text)
print(f"Original: {text}")
print(f"Replaced: {result}")

# Another example - remove the numbers completely
result2 = re.sub(pattern, r'\1', text)
print(f"Numbers removed: {result2}")

# Example - keep only the numbers
result3 = re.sub(pattern, r'\2', text)
print(f"Only numbers: {result3}")

# Example - surround the pattern with brackets
result4 = re.sub(pattern, r'[\1\2]', text)
print(f"With brackets: {result4}")

# Example with a more specific pattern - replace "apple" followed by numbers
specific_result = re.sub(r'(apple)(\d+)', r'\1_MODIFIED_\2', text)
print(f"Specific replacement: {specific_result}")
```
