```python
from datetime import datetime
import pytz

def convert_ist_to_pst_est_cst(ist_time_str):
    # Define the time zones
    ist = pytz.timezone('Asia/Kolkata')
    pst = pytz.timezone('America/Los_Angeles')
    est = pytz.timezone('America/New_York')
    cst = pytz.timezone('America/Chicago')

    # Parse the input IST time string
    ist_time = ist.localize(datetime.strptime(ist_time_str, '%Y-%m-%d %H:%M:%S'))

    # Convert to PST and EST
    pst_time = ist_time.astimezone(pst)
    est_time = ist_time.astimezone(est)
    cst_time = ist_time.astimezone(cst)

    return ist_time, pst_time.strftime('%Y-%m-%d %H:%M:%S'), est_time.strftime('%Y-%m-%d %H:%M:%S'), cst_time.strftime('%Y-%m-%d %H:%M:%S')

def convert_pst_to_ist_est_cst(pst_time_str):
    # Define the time zones
    ist = pytz.timezone('Asia/Kolkata')
    pst = pytz.timezone('America/Los_Angeles')
    est = pytz.timezone('America/New_York')
    cst = pytz.timezone('America/Chicago')

    # Parse the input IST time string
    pst_time = pst.localize(datetime.strptime(pst_time_str, '%Y-%m-%d %H:%M:%S'))

    # Convert to PST and EST
    ist_time = pst_time.astimezone(ist)
    est_time = pst_time.astimezone(est)
    cst_time = pst_time.astimezone(cst)

    return pst_time,ist_time.strftime('%Y-%m-%d %H:%M:%S'), est_time.strftime('%Y-%m-%d %H:%M:%S'), cst_time.strftime('%Y-%m-%d %H:%M:%S')


def convert_est_to_ist_pst_cst(est_time_str):
    # Define the time zones
    ist = pytz.timezone('Asia/Kolkata')
    pst = pytz.timezone('America/Los_Angeles')
    est = pytz.timezone('America/New_York')
    cst = pytz.timezone('America/Chicago')

    # Parse the input IST time string
    est_time = est.localize(datetime.strptime(est_time_str, '%Y-%m-%d %H:%M:%S'))

    # Convert to PST and EST
    ist_time = est_time.astimezone(ist)
    pst_time = est_time.astimezone(pst)
    cst_time = est_time.astimezone(cst)

    return est_time_str,ist_time.strftime('%Y-%m-%d %H:%M:%S'), pst_time.strftime('%Y-%m-%d %H:%M:%S'), cst_time.strftime('%Y-%m-%d %H:%M:%S')

def convert_cst_to_ist_est_pst(cst_time_str):
    # Define the time zones
    ist = pytz.timezone('Asia/Kolkata')
    pst = pytz.timezone('America/Los_Angeles')
    est = pytz.timezone('America/New_York')
    cst = pytz.timezone('America/Chicago')

    # Parse the input IST time string
    cst_time = cst.localize(datetime.strptime(cst_time_str, '%Y-%m-%d %H:%M:%S'))

    # Convert to PST and EST
    ist_time = cst_time.astimezone(ist)
    pst_time = cst_time.astimezone(pst)
    est_time = cst_time.astimezone(est)

    return cst_time_str,ist_time.strftime('%Y-%m-%d %H:%M:%S'), pst_time.strftime('%Y-%m-%d %H:%M:%S'), est_time.strftime('%Y-%m-%d %H:%M:%S')

```

```python
# Example usage
ist_time_input = '2024-08-15 19:00:00'
ist_time, pst_time, est_time, cst_time = convert_ist_to_pst_est_cst(ist_time_input)


print(f'IST (Harsha, Aswathy): {ist_time}, PST (Aaron): {pst_time}, EST (Murali): {est_time}, CST (Jackie): {cst_time}')
```
